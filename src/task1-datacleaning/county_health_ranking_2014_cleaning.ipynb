```python
# Drop all the columns containing quartile ,CI , and '.1'

# Define the function to drop columns containing 'CI', 'Quartile', and '.1'
def drop_unwanted_columns(df):
    # Identify columns that contain 'CI', 'Quartile', or '.1'
    columns_to_drop = [col for col in df.columns if 'CI' in col or 'Quartile' in col or '.1' in col]
    # Drop the identified columns from the dataframe
    df.drop(columns=columns_to_drop, inplace=True)
    return df

# Assume 'ranked_measure_data_df' is your loaded DataFrame
# Apply the updated function to the dataframe
rankedcleaned_data = drop_unwanted_columns(ranked_measure_data_df)

```

```python
# Columns to keep

columns_to_keep = [
    'FIPS', 'State', 'County', 'Deaths', 'YPLL Rate', '% Fair/Poor',
    'Physically Unhealthy Days', 'Mentally Unhealthy Days', '% LBW',
    '% Smokers', '% Obese', 'Food Environment Index', '% Physically Inactive',
    '% With Access', '% Excessive Drinking', '% Alcohol-Impaired',
    'Chlamydia Rate', 'Teen Birth Rate', '% Uninsured', 'PCP Rate',
    'PCP Ratio', 'Dentist Rate', 'Dentist Ratio', 'MHP Rate', 'MHP Ratio',
    'ACSC Rate', '% HbA1c', '% Mammography', 'Graduation Rate',
    '% Some College', '% Unemployed', '% Children in Poverty',
    '% Single-Parent Households', 'Violent Crime Rate', 'Injury Death Rate',
    'Average daily PM25', '% Severe Housing Problems'
]

# Assuming 'data' is your DataFrame
# You can then drop the columns not in 'columns_to_keep' list as follows:

rankedcleaned_data = rankedcleaned_data[columns_to_keep]
```

```python
# Drop unwanted columns containing 'CI' from additional measure data

def unwanted_ci_columns(df):
    # Identify columns that contain 'CI'
    columns_to_drop = [col for col in df.columns if 'CI' in col]
    
    # Drop the identified columns from the dataframe
    df.drop(columns=columns_to_drop, inplace=True)
    
    return df

measured_cleaned_data = unwanted_ci_columns(additional_measure_data_df)
```

```python
# Verify columns have been dropped
# Ignore demographic information because we already have it from census
measured_cleaned_data.columns
```

```python
# Drop all demographic information as we already have it from census
demographic_columns = ['Population', '< 18', '65 and over',
       'African American', 'American Indian/ Alaskan Native', 'Asian',
       'Native Hawaiian/ Other Pacific Islander', 'Hispanic',
       'Non-Hispanic white', '# Not Proficient in English',
       '% Not Proficient in English', 'Female', 'Rural']

measured_cleaned_data.drop(columns=demographic_columns, axis=1, inplace=True)
```

```python
# Inspect remaining columns
measured_cleaned_data.columns
```

```python
# Keep the only columns we need
columns_to_keep = ['FIPS', 'State', 'County','% Diabetic','HIV Rate',
                  'Age-adjusted Mortality', 'Infant mortality rate',
                  'Child mortality rate', '% Food Insecure','% Limited Access',
                  'Costs','Household Income','% Free Lunch']

measured_cleaned_data = measured_cleaned_data[columns_to_keep]
```

```python
measured_cleaned_data.columns
```

```python
# Combine both of those dataset together

# Merging the DataFrames on 'FIPS', 'State', and 'County'
merged_data = pd.merge(rankedcleaned_data, measured_cleaned_data, 
                       on=['FIPS', 'State', 'County'], 
                       how='inner')
```

```python
merged_data.columns
```

```python
# Create a column to indicate the year of the dataset and records
# All the year value here will be 2014 since this is 2014 data
merged_data['Year'] = '2014'
```

```python
# Convert it to a csv file
merged_data.isnull().mean()
```

```python
# Convert it to a csv file
merged_data.to_csv('CountyRanking2014.csv', index=False)
```
