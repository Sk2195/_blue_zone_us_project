```python
# Create a new list
df_list = [df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]

# Initialize merged_df with the first dataframe
merged_df = df_list[0]

# Merge all dataframes based on common columns State and County
for i, df in enumerate(df_list[1:], start=2):
    suffix = f'_{i}'
    merged_df = pd.merge(merged_df, df, on=["State", "County"], suffixes=('', suffix))
```

```python
# Define the suffixes for the columns to sum
suffixes = [''] + [f'_{i}' for i in range(2, len(df_list) + 1)]
print(suffixes)
```

```python
# Create a list of columns for which the values should be added
cols_to_add = ["Days with AQI","Good Days","Moderate Days","Unhealthy for Sensitive Groups Days",
              "Unhealthy Days","Very Unhealthy Days", "Hazardous Days","Days CO", "Days NO2",
              "Days Ozone", "Days PM2.5", "Days PM10"]

for column in cols_to_add:
  # Sum the columns for 'column' using filter
  merged_df[column] = merged_df.filter(like=column).sum(axis=1)

  # Drop only the individual columns used for summation
  merged_df = merged_df.drop(columns=[f"{column}{x}" for x in suffixes[1:]])
```

```python
# Create a new list of required columns
cols_to_modify = ["Max AQI", "90th Percentile AQI", "Median AQI"]
for column in cols_to_modify:

  max_aqi_columns = merged_df.filter(like=column)
  # Get the maximum value across all columns for each row
  merged_df[column] = max_aqi_columns.max(axis=1)

  # Drop the individual "Median AQI" columns
  merged_df = merged_df.drop(columns=[f"{column}{x}" for x in suffixes[1:]])
```

```python
# Drop the year columns
merged_df.drop(columns= ['Year'],inplace=True)
merged_df = merged_df.drop(columns=[f"Year{x}" for x in suffixes[1:]])
```
